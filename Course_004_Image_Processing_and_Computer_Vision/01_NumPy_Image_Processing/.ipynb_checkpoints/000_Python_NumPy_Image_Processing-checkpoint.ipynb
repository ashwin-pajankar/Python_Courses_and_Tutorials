# Image Mask

```python
img1
```

```python
nrows, ncols, channels = img1.shape
row, col = np.ogrid[:nrows, :ncols]
cnt_row, cnt_col = nrows/2, ncols/2
```

```python
outer_disk_mask = (((row - cnt_row)**2 + (col - cnt_col)**2) > (nrows/2))
img2 = img1.copy()
img2[outer_disk_mask] = [0, 0, 0] 
```

```python
plt.imshow(img2)
plt.axis('off')
plt.title('Masked Image')
plt.show()
```

# Image Channels

```python
# Color image
img1 = plt.imread('D:/Dataset/4.1.04.tiff')
```

```python
r = img1[:, :, 0]
g = img1[:, :, 1]
b = img1[:, :, 2]
```

```python
type(r)
```

```python
r.shape
```

```python
r.ndim
```

```python
output = [img1, r, g, b]
titles = ['Original', 'Red', 'Green', 'Blue']
```

```python
for i in range(4):
    plt.subplot(2, 2, i+1)
    
    plt.axis('off')
    plt.title(titles[i])
    if i == 0:
        plt.imshow(output[i])
    else:
        plt.imshow(output[i], cmap='gray')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.axis('off')
plt.title(titles[0])
plt.imshow(output[0])

plt.subplot(2, 2, 2)
plt.axis('off')
plt.title(titles[1])
plt.imshow(output[1], cmap='Reds')

plt.subplot(2, 2, 3)
plt.axis('off')
plt.title(titles[2])
plt.imshow(output[2], cmap='Greens')

plt.subplot(2, 2, 4)
plt.axis('off')
plt.title(titles[3])
plt.imshow(output[3], cmap='Blues')
```
