```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "D:\\Dataset\\"
imgpath1 =  path + "4.2.01.tiff"
imgpath2 =  path + "4.2.03.tiff"
```

```python
img1 = cv2.imread(imgpath1, 1)
img2 = cv2.imread(imgpath2, 1)
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
x = np.uint8([240])
y = np.uint8([20])
```

```python
print(x + y) # (x + y) % 255
print(cv2.add(x, y)) # 255
```

```python
add1 = img1 + img2
add2 = cv2.add(img1, img2)
```

```python
titles = ['Liquid Drop', 'Mandrill', 
          'NumPy Addition', 'cv2.add()']
```

```python
images = [img1, img2, add1, add2]
```

```python
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
```
