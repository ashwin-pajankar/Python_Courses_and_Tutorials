```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "D:\\Dataset\\"
imgpath1 =  path + "5.1.11.tiff"

img = cv2.imread(imgpath1, 0)

th = 0
max_val = 255
ret, binary_inv = cv2.threshold(img, th, max_val, 
                                cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU )
plt.imshow(binary_inv, cmap='gray')
plt.title('Thresholded Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#k = np.ones((5, 5), np.uint8)
#k = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
#k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5, 5))
k = cv2.getStructuringElement(cv2.MORPH_CROSS,(5, 5))
print(k)
```

```python
erosion = cv2.erode(binary_inv, k, iterations = 1)
plt.imshow(erosion, cmap='gray')
plt.title('Eroded Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
dilation = cv2.dilate(binary_inv, k , iterations = 1)
plt.imshow(dilation, cmap='gray')
plt.title('Dilated Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# erosion followed by dilation
opening = cv2.morphologyEx(binary_inv, cv2.MORPH_OPEN, k)
plt.imshow(opening, cmap='gray')
plt.title('Opening')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# Dilation followed by Erosion
closing = cv2.morphologyEx(binary_inv, cv2.MORPH_CLOSE, k)
plt.imshow(closing, cmap='gray')
plt.title('Closing')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# difference between dilation and erosion of an image
gradient = cv2.morphologyEx(binary_inv, cv2.MORPH_GRADIENT, k)
plt.imshow(gradient, cmap='gray')
plt.title('Gradient')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# difference between input image and Opening of the image
tophat = cv2.morphologyEx(binary_inv, cv2.MORPH_TOPHAT, k)
plt.imshow(tophat, cmap='gray')
plt.title('Tophat')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# difference between the closing of the input image and input image
blackhat = cv2.morphologyEx(binary_inv, cv2.MORPH_BLACKHAT, k)
plt.imshow(blackhat, cmap='gray')
plt.title('Blackhat')
plt.xticks([])
plt.yticks([])
plt.show()
```
