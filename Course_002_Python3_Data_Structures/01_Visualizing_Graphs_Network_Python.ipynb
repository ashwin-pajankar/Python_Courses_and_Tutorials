```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
```

# Creating Graphs and Networks

```python
G = nx.Graph()
```

```python
type(G)
```

```python
print(G.nodes())
print(G.edges())
print(type(G.nodes()))
print(type(G.edges()))
```

```python
G.add_node('a')
```

```python
G.add_nodes_from(['b', 'c'])
```

```python
print(G.nodes())
```

```python
print(G.edges())
```

```python
G.add_edge(1, 2)
```

```python
edge = ('d', 'e')
G.add_edge(*edge)
```

```python
edge = ('a', 'b')
G.add_edge(*edge)
```

```python
print(G.nodes())
```

```python
print(G.edges())
```

```python
G.add_edges_from([('a', 'c'), ('c', 'd'),
                 ('a', 1), (1, 'd'),
                 ('a', 2)])
```

```python
print(G.nodes())
```

```python
print(G.edges())
```

# Visualizing Graphs and Networks

```python
nx.draw(G)
plt.show()
```

```python
nx.draw(G)
plt.show()
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
nx.draw(G, with_labels=True,
       font_weight='bold')
plt.show()
```

# Petersen Graph

```python
G = nx.petersen_graph()
```

```python
G.nodes()
```

```python
G.edges()
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
nx.draw_shell(G, nlist=[range(5, 10),
                       range(5)],
             with_labels=True,
             font_weight='bold')
plt.show()
```

```python
options = {'node_color': 'black',
          'node_size': 100,
          'width': 3}
nx.draw_random(G, **options)
plt.show()
```

```python
nx.draw_circular(G, **options)
plt.show()
```

```python
nx.draw_spectral(G, **options)
plt.show()
```

```python
nx.draw_shell(G, nlist=[range(5, 10),
                       range(5)],
             **options)
plt.show()
```

```python
G = nx.dodecahedral_graph()
```

```python
G.nodes()
```

```python
G.edges()
```

```python
shells=[[2, 3, 4, 5, 6],
       [8, 1, 0, 19, 18,
       17, 16, 15, 14, 7],
       [9, 10, 11, 12, 13]]
nx.draw_shell(G, nlist=shells, **options)
plt.show()
```

# Assign Custom labels to nodes

```python
G = nx.path_graph(4)
```

```python
cities = {0: 'Mumbai',
          1: 'Hyderabad',
          2: 'Bangalore',
          3: 'New York'}
H=nx.relabel_nodes(G, cities)
```

```python
H.edges()
```

```python
H.nodes()
```

```python
nx.draw(H, **options)
plt.show()
```

```python
nx.draw(H, with_labels=True)
plt.show()
```
