# Fast Fourier Transform

```python
%matplotlib inline
```

```python
# 1-D FFT

t = np.arange(128)
print(t)
```

```python
sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(t.shape[-1])
```

```python
print(sp, freq)
```

```python
plt.subplots_adjust(hspace=0.5)

plt.subplot(2, 1, 1)
plt.title('Real Part')
plt.plot(freq, sp.real)

plt.subplot(2, 1, 2)
plt.title('Imaginary Part')
plt.plot(freq, sp.imag)

plt.show()
```

```python
# Inverse 1-D DFT
t = np.arange(500)
n = np.zeros((500, ), dtype = complex)
```

```python
n[40:80] = np.exp(1j*np.random.uniform(0, 2*np.pi, (40, )))
s = np.fft.ifft(n)
```

```python
plt.plot(t, s.real)
plt.show()
```

```python
plt.plot(t, s.imag)
plt.show()
```

### More complex example of an N-D FFT

```python
[X, Y] = np.meshgrid(2 * np.pi * np.arange(200) / 12,
                     2 * np.pi * np.arange(200) / 34)
print(X, Y)
```

```python
plt.plot(X, Y)
plt.show()
```

```python
n =  np.sin(X) +np.cos(Y) + np.random.uniform(0, 1, X.shape)
print(n)
```

```python
plt.imshow(n)
plt.show()
```

```python
z = np.fft.fftn(n).astype(np.uint8)
z
```

```python
plt.imshow(n, cmap='hot')
plt.show()
```

```python
plt.imshow(n, cmap='cool')
plt.show()
```

### Inverse FFT

```python
n = np.zeros((200, 200), dtype=complex)
n[60:80, 20:40] = np.exp(1j*np.random.uniform(0, 2 * np.pi, (20, 20)))
```

```python
n
```

```python
z = np.fft.ifftn(n)
z
```

```python
plt.imshow(z.real, cmap='cool')
plt.show()
```

```python
plt.imshow(z.imag, cmap='cool')
plt.show()
```

# Linear Algebra

## Dot Product

```python
np.dot(10, 5)
```

```python
a = np.array([2, 4], np.uint8)
b = np.array([5, 6], np.uint8)

np.dot(a, b)
```

```python
a = np.array([[2, 4], [5, 8]], np.uint8)
b = np.array([[5, 6], [9, 3]], np.uint8)

np.dot(a, b)
```

```python
a = np.random.rand(10, 5)
b = np.random.rand(5, 2)
c = np.random.rand(2, 3)
d = np.random.rand(3, 4)

e = np.linalg.multi_dot([a, b, c, d])
print(e)
```

## Vector Dot Product

```python
a = np.array([2, 4], np.uint8)
b = np.array([5, 6], np.uint8)

np.vdot(a, b)
```

```python
a = np.array([2 + 4j, 5 + 8j])
b = np.array([5 + 6j, 9 + 3j])

np.vdot(a, b)
```

## Inner Products

```python
a = np.array([1, 2, 3, 4, 5, 6])
b = np.array([0, 1, 0, 1, 0, 1])

np.inner(a, b)
```

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([[0, 1, 0], [1, 0, 1]])

np.inner(a, b)
```
